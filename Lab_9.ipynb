# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
import collections


def get_user_graph():
    graph = collections.defaultdict(dict)
    num_nodes = int(input("Enter the number of nodes in the graph: "))
    nodes = []
    for i in range(num_nodes):
        node_name = input(f"Enter name for node {i+1}: ").strip().upper()
        nodes.append(node_name)

    print("\nEnter edges and weights. Enter 'done' when finished.")
    while True:
        edge_input = input("Enter edge (e.g., A B 5 for A to B with weight 5) or 'done': ").strip().split()
        if edge_input[0].lower() == 'done':
            break
        if len(edge_input) == 3:
            u, v, weight_str = edge_input[0].upper(), edge_input[1].upper(), edge_input[2]
            try:
                weight = int(weight_str)
                if u in nodes and v in nodes:
                    graph[u][v] = weight
                else:
                    print(f"Nodes {u} or {v} not defined in your initial node list. Please ensure all nodes are listed first.")
            except ValueError:
                print("Invalid weight. Please enter an integer.")
        else:
            print("Invalid input format. Please try again (e.g., A B 5).")
    return dict(graph)

def dijkstra_with_path(graph, start):
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph}
    distances[start] = 0
    unvisited = set(graph.keys())

    while unvisited:

        min_node = None
        for node in unvisited:
            if min_node is None or distances[node] < distances[min_node]:
                min_node = node

        if min_node is None or distances[min_node] == float('inf'):
            break

        unvisited.remove(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node
    return distances, predecessors


def print_shortest_path(start, target, distances, predecessors):
    if distances[target] == float('inf'):
        print(f"No path found from {start} to {target}. (Distance: Unreachable)")
        return

    path = []
    current = target
    while current is not None and current != start:
        path.insert(0, current)
        current = predecessors[current]

    if current == start:
        path.insert(0, start)
        print(f"Shortest path from {start} to {target}: {' -> '.join(path)} (Distance: {distances[target]})")
    else:
        print(f"Could not reconstruct path from {start} to {target}.")


print("\n--- Using an expanded example graph for demonstration ---")
expanded_graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 7},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 6},
    'E': {'B': 7, 'D': 6, 'F': 5},
    'F': {'C': 8, 'E': 5}
}
current_graph = expanded_graph
print("Graph:", current_graph)


source_nodes_to_test = ['A', 'E', 'Z']

for start_node in source_nodes_to_test:
    print(f"\n--- Testing Dijkstra from source node: {start_node} ---")
    if start_node not in current_graph:
        print(f"Source node '{start_node}' not found in the graph. Skipping.")
        continue

    distances, predecessors = dijkstra_with_path(current_graph, start_node)

    print(f"\nShortest Distances from {start_node}:")
    for node, dist in distances.items():
        if dist == float('inf'):
            print(f"  To {node}: Unreachable")
        else:
            print(f"  To {node}: {dist}")

    print(f"\nShortest Paths from {start_node}:")
    for node in current_graph:

        print_shortest_path(start_node, node, distances, predecessors)

```
